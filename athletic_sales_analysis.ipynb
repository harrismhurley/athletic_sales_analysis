```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv("./Resources/athletic_sales_2020.csv")
sales_2021_df = pd.read_csv("./Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales_20_21 = pd.concat([sales_2020_df, sales_2021_df], axis="rows", join="inner").reset_index(drop=True)
sales_20_21.head()
```

```python
# Check if any values are null.
null_counts = sales_20_21.isnull().sum()
print(null_counts)

```

```python
# Check the data type of each column
sales_20_21.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
sales_20_21["invoice_date"] = pd.to_datetime(sales_20_21["invoice_date"])
sales_20_21.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_20_21.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
products_sold_summary = (
    sales_20_21.groupby(['region', 'state', 'city'])['units_sold']
    .sum()
    .to_frame(name='Total_Products_Sold')
    .sort_values(by='Total_Products_Sold', ascending=False)
)
# Show the top 5 results.
products_sold_summary.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold_pivot = sales_20_21.pivot_table(
    values="units_sold",
    index=["region", "state", "city"],
    aggfunc="sum",
)

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)


# Show the top 5 results.
products_sold_pivot.sort_values(by='Total_Products_Sold', ascending=False).head()

```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
sales_20_21.groupby(['region', 'state', 'city'])['total_sales'].sum() \
            .reset_index() \
            .rename(columns={'total_sales': 'Total Sales'}) \
            .sort_values(by="Total Sales", ascending=False) \
            .head()
            


# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_20_21.pivot_table(
    values='total_sales', 
    index=['region', 'state', 'city'], 
    aggfunc='sum'
# Optional: Rename the "total_sales" column to "Total Sales"
) \
    .rename(columns={'total_sales': 'Total Sales'}) \
    .sort_values(by="Total Sales", ascending=False) \
    .head()

# Show the top 5 results.
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
sales_20_21.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum() \
            .rename('Total Sales') \
            .sort_values(ascending=False) \
            .head()

# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
sales_20_21.pivot_table(
    values='total_sales', 
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum'
) \
    .rename('Total Sales') \
    .sort_values(ascending=False) \
    .head()

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear = sales_20_21[sales_20_21['product'] == "Women's Athletic Footwear"]
womens_athletic_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
sales_20_21[sales_20_21['product'] == "Women's Athletic Footwear"] \
    .groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum() \
    .rename('Womens_Footwear_Units_Sold') \
    .sort_values(ascending=False) \
    .head()
# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
sales_20_21[sales_20_21['product'] == "Women's Athletic Footwear"] \
    .pivot_table(
        values='units_sold', 
        index=['retailer', 'region', 'state', 'city'], 
        aggfunc='sum'
    ) \
    .rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}) \
    .sort_values(by='Womens_Footwear_Units_Sold', ascending=False) \
    .head()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"


# Show the top 5 results.
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_20_21[sales_20_21['product'] == "Women's Athletic Footwear"] \
    .pivot_table(
        values='total_sales', 
        index='invoice_date', 
        aggfunc='sum'
    ) \
    .rename(columns={'total_sales': 'Total Sales'}) \
    .head(10)

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
sales_20_21[sales_20_21['product'] == "Women's Athletic Footwear"] \
    .pivot_table(
        values='total_sales', 
        index='invoice_date', 
        aggfunc='sum'
    ) \
    .resample('D').sum() \
    .sort_values(by='total_sales', ascending=False)

# Sort the resampled pivot table in descending order on "Total Sales".
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
sales_20_21[sales_20_21['product'] == "Women's Athletic Footwear"] \
   .pivot_table(
       values='total_sales', 
       index='invoice_date', 
       aggfunc='sum'
   ) \
   .resample('W').sum() \
   .sort_values(by='total_sales', ascending=False)

# Sort the resampled pivot table in descending order on "Total Sales".
```
